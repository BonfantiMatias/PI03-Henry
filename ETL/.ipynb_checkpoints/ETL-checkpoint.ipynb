```python
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
from sqlalchemy import create_engine


```

```python
#Transformacion de csv a dataframe pandas
df = pd.read_csv('traducido.csv')
df.head()
```

```python
#Reemplazar los signos '?' por espacios vacios
#Definimos una funcion

def formateo(x):
    
    if x =='?':
        x = ''
    return x

#Automatizamos el proceso para todas las columnas 
columnas = df.columns
for columna in columnas:
    df[columna] = df[columna].apply(formateo)
```

```python
#Normalizar la columna Ruta
df = df.replace({'Ruta': ' ' }, {'Ruta': ''}, regex=True)
df = df.replace({'Ruta': ',' }, {'Ruta': ', '}, regex=True)
df
```

```python
#Cargamos Dataset con coordenadas de Accidente
coordenadas = pd.read_csv('coordenadas.csv')
coordenadas.shape
```

```python
#Concatenamos los 2 Dataset
df = pd.concat([df, coordenadas], axis="columns", join="inner")
df.shape
```

```python
df.columns
```

```python
df = df[['fecha', 'HORA declarada', 'Ruta', 'lat', 'lon',
       'OperadOR', 'flight_no', 'route', 'ac_type', 'registration', 'cn_ln',
       'all_aboard', 'PASAJEROS A BORDO', 'crew_aboard',
       'cantidad de fallecidos', 'passenger_fatalities', 'crew_fatalities',
       'ground', 'summary', 'nueva', 'reporte']]
```

```python
#Separar los fabricantes de los modelos de aviones
aviones = df["ac_type"].str.split(" ", n = 1, expand = True)
```

```python
#Detalle
aviones.sample(5)
```

```python
#Verificar cantidades de Empresas
aviones.groupby(0).size().sort_values()
```

```python
#Corregir Havilland
def Havi(x):
    if x =='de':
        x = 'Havilland'
    elif x =='De':
        x = 'Havilland'
    return x

aviones[0] = aviones[0].apply(Havi)
aviones = aviones.replace({1: 'Havilland' }, {1: ''}, regex=True)
aviones = aviones.replace({1: 'havilland' }, {1: ''}, regex=True)
```

```python
aviones.shape
```

```python
#Unir Dataframe completo con aviones
df =pd.concat([df, aviones], axis="columns", join="inner")
```

```python
#Tipos de datos
df.dtypes
```

```python
#Definir columnas del tipo integer
enteros = ['all_aboard', 'PASAJEROS A BORDO', 'crew_aboard',
       'cantidad de fallecidos', 'passenger_fatalities', 'crew_fatalities',
       'ground']
def entero(x):
    if x =='':
        x = 0
    return x
for colum in enteros:
    df[colum] = df[colum].apply(entero)
    df[colum] = df[colum].astype('int32')
```

```python
#Listar columnas
df.columns
```

```python
#Renombrar columnas
df = df.rename(columns={'fecha':'Fecha Accidente', 'HORA declarada':'Hora Accidente', 'Ruta':'Lugar del Accidente', 'OperadOR':'Operador Aereo', 'flight_no': 'Nro de vuelo', 'route':'Ruta de vuelo', 'ac_type':'Modelo', 'registration':'Registro Avion', 
                        'all_aboard':'Total Personas a bordo', 'PASAJEROS A BORDO':'Pasajeros', 'crew_aboard':'Tripulantcion', 'cantidad de fallecidos':'Total de fallecidos', 'passenger_fatalities':'Pasajeros Fallecidos', 'crew_fatalities':'Tripulantes Fallecidos','ground':'Personas en tierra fallecidas', 'reporte': 'Reporte de Accidente',
                       0:'Constructor Avion',1:'Modelo Avion'})
```

```python
#Eliminar Columnas
df = df.drop(['summary','nueva','Modelo'], axis='columns')
```
